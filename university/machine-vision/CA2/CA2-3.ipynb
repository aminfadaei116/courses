```
## The imports 
import numpy as np
from PIL import Image 
import cv2
from matplotlib import pyplot as plt
import matplotlib.lines as mlines
from numpy import random
from sklearn.preprocessing import normalize
```

```
#define Variables
Train = []
Test = []


for i in range(25):
    Train.append([[] for _ in range(15)])
    Test.append([[] for _ in range(15)])
PIC_PATH = "/content/drive/MyDrive/Machine Vision/HW2/cropped_faces"
```

```
for i in range(25):
  for j in range(15):
    Train[i][j] = cv2.imread(PIC_PATH + "/s" + "{0:0=2d}".format(i+1) + "_" + "{0:0=2d}".format(j+1) + ".jpg", cv2.IMREAD_COLOR) 

for i in range(25):
  for j in range(15):
    Test[i][j] = cv2.imread(PIC_PATH + "/s" + "{0:0=2d}".format(i+1) + "_" + "{0:0=2d}".format(j+1) + ".jpg", cv2.IMREAD_COLOR) 
```

```
SIFT = cv2.SIFT_create()
BF = cv2.BFMatcher(cv2.NORM_L2, crossCheck=False)
```

```
Image_matrix2 = np.zeros((25 * 15, 25 * 15))
Threshold = 0.65
for i in range(25): #In this part we are going to make a matrix that will show the distance betwee images
  for j in range(15):
    _, desc1 = SIFT.detectAndCompute(Train[i][j], mask = None)
    desc1 = normalize(desc1)
    size1 = desc1.shape[0]
    for next_i in range(25):
      for next_j in range(15):

        if next_i == i and next_j == j:
          continue
        _, desc2 = SIFT.detectAndCompute(Train[next_i][next_j], mask = None)
        desc2 = normalize(desc2)
        size2 = desc2.shape[0]
        # KNN
        rawMatches = BF.knnMatch(desc1, desc2, 2)
        number_match = 0
        for m, n in rawMatches:
          if m.distance < n.distance * Threshold:
            number_match += 1
        Image_matrix2[15*i+j][15*next_i+next_j] = number_match/((size1))
    print((15 * i + j) / (25 * 15))

```

```
Image_matrix = np.zeros((25 * 15, 25 * 15))
Threshold = 0.65
for i in range(25): #In this part we are going to make a matrix that will show the distance betwee images
  for j in range(15):
    _, desc1 = SIFT.detectAndCompute(Test[i][j], mask = None)
    desc1 = normalize(desc1)
    size1 = desc1.shape[0]
    for next_i in range(25):
      for next_j in range(15):

        if next_i == i and next_j == j:
          continue
        _, desc2 = SIFT.detectAndCompute(Test[next_i][next_j], mask = None)
        desc2 = normalize(desc2)
        size2 = desc2.shape[0]
        # KNN
        rawMatches = BF.knnMatch(desc1, desc2, 2)
        number_match = 0
        for m, n in rawMatches:
          if m.distance < n.distance * Threshold:
            number_match += 1
        Image_matrix[15*i+j][15*next_i+next_j] = number_match/((size1))
    print((15 * i + j) / (25 * 15))

```

```
arr = []
full_arr = []
correct = 0
acc = [[] for _ in range(25)]
for i in range(Image_matrix.shape[0]):
  arr = Image_matrix[i].argsort()[-14:][::-1]
  arr = arr / 15
  arr = arr.astype(int)
  full_arr.append(arr)
  truth = (arr == int(i/15))
  correct += sum(truth)
  acc[int(i/15)].append(sum(truth))

print(correct / (Image_matrix.shape[0] * 14))
```
